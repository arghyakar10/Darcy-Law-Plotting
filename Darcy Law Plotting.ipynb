```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
re = 3000
rwf = 0.5
pwf = 1500
h = 60
k = 150
q = 200
meu = 15
B = 1
```

```python
r = np.linspace(rwf,re,500)
```

```python
r
```

```python
pressure = []
for i in range(len(r)):
    p = pwf + (141.2*q*B*meu*(np.log(r[i]/rwf)))/(k*h)
    pressure.append(p)
```

```python
pressure
```

```python
plt.plot(r,pressure)
```

```python
plt.plot(r,pressure)
plt.xlabel('x(ft)')
plt.ylabel('Pressure')
plt.grid(True)
plt.title('Reservoir Pressure Profile')
```
